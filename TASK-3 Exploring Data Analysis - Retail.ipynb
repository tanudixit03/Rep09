<font size=6><b>Data Visualisation</b>

<B>correlation matrix using heatmap to check pairwise dependency<B/>

```python
plt.figure(figsize=(7,4))
sns.heatmap(df.corr(), cmap = 'viridis', annot = True, linewidth = 4)


#annot = True: write the data value in each cell
```

<B>From the above heat map we can infer that:-</b>
    
    *correlation between Sales and Profit is positive and moderate
    *correlation between Qunatity and Discount is positive but low 
    *correlation between Profit and Discount as well as Sales and Discount are negatively correlated.

```python
# Use of pairplot to show pairwise relationship in the Dataset
sns.pairplot(df,diag_kind='data')
```

<b>Observations:</b>

*Sales increases profit also increases

*As we increase the Discount profit strarts decreasing ( Discount is more than 30%, Profit decreases)

*As Discount increases, Sales decreases

<font size=4><b>Now we will do Ship Mode wise Sales Analysis</b>

```python
df['Ship Mode'].value_counts()
#it will return the unique values in the dataframe
```

```python
# Used a countplot for Ship Mode
plt.figure(figsize=(6,4))
sns.countplot(x=df["Ship Mode"])
plt.show()
```

Standard class most opted ship mode and Same Day is the least opted.

<font size=5><b> Region wise Sales Analysis</b>

```python
df['Region'].value_counts()
# returns count of unique value from the data
```

```python
#using Barplot to show profits across different regions
plt.figure(figsize = (6,4))
sns.set(font_scale=1)
sns.barplot(x='Region', y='Profit', data=df, hue='Segment')
plt.title("Profit across different Regions(by Segments)", fontsize = 15)
plt.show()
```

```python
# Using pie chart to show region sales
df['Region'].value_counts().plot(kind="pie",figsize=(5,5),autopct="%1.1f%%",shadow="True")
plt.legend(fontsize=10)
plt.show
```

<b>observation</b>:- Maximum no. of sales are in western region.
Minimum no. of sales in southern region.

<font size=5><b> Region wise Ship Mode Analysis</b>

```python
plt.figure(figsize=(7,6))
sns.countplot(x="Region", hue="Ship Mode", data=df,palette='magma')
plt.ylabel("Ship Mode Count")
plt.show()
```

<b>Observation</b>:-
 Standard Class is most opted Ship Mode and Same Day is least opted Ship Mode among different regions.

<font size=5><b><u>Sales and Profit distribution among different Regions.</b></u>

```python
sp = df.groupby("Region")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"],ascending=False)
sp.plot(kind="bar", fontsize=10, color=["purple","pink"])
plt.xlabel("Region", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)
plt.show()
```

<font size=5><b>Segment Wise Sales Analysis</b>

```python
df['Segment'].value_counts()
#count of the values in the dataset.
```

```python
df['Segment'].value_counts().plot(kind="pie",figsize=(5,5),autopct="%1.1f%%",fontsize = 15, colors=["green","red","royalblue"],shadow="True")
plt.legend()
plt.show()
```

<font size=5><b><u>Citywise Sales Analysis</b>

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=df['City'], order=(df['City'].value_counts().head(60)).index)
plt.xticks(rotation=60)
plt.xlabel('City', fontsize = 15)
plt.ylabel('Count / No.of buyers', fontsize = 15)
plt.show()
```

<b>Observation:</b>
New York City has highest count of buyers followed by Los Angeles and then by Philadelphia.

<Font size=5><b>Statewise Sales Analysis</b>

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=df['State'], order=(df['State'].value_counts().head(60)).index)
plt.xticks(rotation=60)
plt.xlabel('States', fontsize = 15)
plt.ylabel('Count / No.of buyers', fontsize = 15)
plt.show()
```

<b>Observation:</b>
California, New York, Texas has highest count of buyers and Wyoming has the least count.

<font size=5><b>Category wise Analysis</b>

```python
df['Category'].value_counts()
```

```python
sns.barplot(x='Category', y='Sales', data=df, palette="magma")
plt.show()
```

<B>Observation:</b>
Sales are mostly in Technology and Furniture Category.

```python
#Barplot for profits corresponding to each category
sns.barplot(x='Category', y='Profit', data=df, palette="magma")
plt.show()
```

<b>Observation</b>:There is more profit in Technology Category and moderate in Furniture and Office Supplies category.

    

```python
csp = df.groupby("Category")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"], ascending = True)
csp.plot(kind="bar",fontsize=10, color = ["green", "red"])
plt.xlabel("Category", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)

plt.show()
```

<b>Observation:-
    
Sales of Office Supplies category is high as well as Profit is more.
    
Sales of Furniture category is high but there is a little Profit.

<font size=5><b>Sales and profit distribution among sub-categories</B></font>

```python
sls = df.groupby("Sub-Category")[["Sales","Profit"]].sum().sort_values(by=["Profit"], ascending = False)
sls.plot(kind="bar", figsize=(8,5), fontsize=10, color = ["orange", "purple"])
plt.xticks(rotation=70)
plt.xlabel("Sub-Category", fontsize=10)
plt.ylabel("Sales and Profit", fontsize=10)
plt.show()
```

<b>Observation</b>:
Papers shows highest profit with moderate sales and
Tables shows least profit with higher sales.
